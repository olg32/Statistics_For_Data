```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Piece_Dim.csv')
```

```python
df.head()
```

```python
sns.relplot(data=df, x='Length', y='Width')
```

```python
sns.relplot(data=df, x='Length', y='Width', hue='Operator')
```

```python
sns.relplot(data=df, x='Length', y='Width', col='Operator')
```

```python
sns.relplot(data=df, x='Item_No', y='Length', kind='line')
```

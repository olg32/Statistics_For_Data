```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Roll a die

```python
one_h = np.random.randint(1,7,100)
one_h
```

```python
sns.displot(x=one_h, bins=6)
```

```python
one_m = np.random.randint(1,7,1000000)
```

```python
sns.displot(x=one_m, bins=6)
```

```python
np.mean(one_m)
```

```python
np.std(one_m)
```

### Obtain a sample

```python
one_h.reshape(25,4).mean(axis=1)
```

```python
mean4 = one_m.reshape(250000,4).mean(axis=1)
```

```python
sns.displot(x=mean4, bins=6)
```

```python
np.mean(mean4)
```

```python
np.std(mean4)
```

```python
mean100 = one_m.reshape(10000,100).mean(axis=1)
```

```python
sns.displot(x=mean100, bins=6)
plt.xlim(0,6)
```

```python
np.mean(mean100)
```

```python
np.std(mean100)
```

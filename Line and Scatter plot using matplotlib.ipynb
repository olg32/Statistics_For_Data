```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Piece_Dim.csv')
```

```python
df.head()
```

```python
plt.plot(df['Length'])
mean_l = df['Length'].mean()
sd_l = df['Length'].std()
plt.axhline(y = mean_l, color = 'red', linestyle = '-')
plt.axhline(y = (mean_l+ 3*sd_l), color = 'red', linestyle = '--')
plt.show()
```

```python
plt.scatter(x=df['Length'], y=df['Width'], marker='v')
plt.show()
```
